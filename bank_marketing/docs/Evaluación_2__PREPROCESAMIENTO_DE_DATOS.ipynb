### 3.2. Limpieza y Transformación de Datos

En esta sección, nos centraremos en refinar nuestros datos para asegurar su calidad y utilidad para el análisis predictivo. Abordaremos dos tareas críticas:

1.  **Manejo de Datos Faltantes (unknown):** Transformaremos los valores 'unknown' a un formato que el modelo pueda interpretar, siguiendo un enfoque ético que evita la imputación arbitraria.
2.  **Manejo de Variables Irrelevantes o con Fuga de Datos (Data Leakage):** Identificaremos y eliminaremos variables que, aunque parezcan útiles, pueden sesgar nuestro modelo o no estarán disponibles en un escenario real de predicción.

#### 3.2.1. Manejo de Datos Faltantes (unknown): "Traduciendo la incertidumbre"

En el sector financiero, la falta de información es tan reveladora como la información misma. Cuando un cliente no revela su nivel educativo o el estado de su crédito, los datos se registraron originalmente como unknown (desconocido).

Para ser transparentes y éticos, no podemos asumir ni inventar información que el cliente no proporcionó. El primer paso analítico es estandarizar esta incertidumbre.

En el siguiente bloque de código, convertimos la etiqueta de texto "unknown" al formato estándar que utiliza Python para representar valores faltantes (NaN). Esto nos permite cuantificar el tamaño de nuestros "puntos ciegos" de manera precisa y evitar que un texto genere sesgos en el futuro algoritmo de machine learning.

```python
# --- FASE 1: ESTANDARIZACIÓN DE VALORES FALTANTES ---

# Definimos las columnas socio-demográficas y financieras clave de nuestros clientes
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan']

# Iteramos (recorremos) cada columna para cambiar la palabra 'unknown' por un valor nulo (NaN).
# Esto le indica a Python que esa casilla está vacía en lugar de tener una respuesta ambigua.
for col in categorical_cols:
    df[col] = df[col].replace('unknown', np.nan)

# --- FASE 2: AUDITORÍA DE RESULTADOS ---

print("--- REPORTE DE DATOS FALTANTES (NaN) ---")
# Contamos cuántos registros quedaron vacíos en cada columna
print(df[categorical_cols].isnull().sum())
```

**Análisis de Datos Faltantes**

El reporte nos muestra exactamente cuántos datos nos faltan en variables críticas para el negocio:

- **Caso de `default` (Historial de Morosidad):**  
  Es la variable con mayor cantidad de datos nulos. Esto refleja la reticencia del cliente a hablar sobre deudas pasadas.

- **Caso de `education` y `job`:**  
  Muestran vacíos que representan a clientes que no desearon compartir esta información.

##### 3.2.1.1. Análisis de Patrones de Nulos: "¿Quiénes ocultan información?"

Para entender si existe un patrón oculto en estos vacíos, analizaremos si los clientes que omiten un dato tienden a omitir otros. En términos de negocio, esto nos permite saber si estamos perdiendo información sobre un perfil de cliente específico.

El siguiente gráfico agrupa las variables según los datos faltantes, permitiéndonos visualizar si existen correlaciones en los datos ausentes.

```python
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Nos aseguramos de que las columnas a evaluar estén definidas
categorical_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan']

# Si el dataframe no está cargado por algún motivo, lo cargamos
if 'df' not in globals() and 'df' not in locals():
    df = pd.read_csv('bank-additional-full.csv', sep=';')
    for col in categorical_cols:
        df[col] = df[col].replace('unknown', np.nan)

# Preparamos la matriz de nulos (1 para nulo, 0 para no nulo)
missing_matrix = df[categorical_cols].isnull().astype(int)

# Tomamos una muestra para que el gráfico sea claro y no se sature
muestra_size = min(500, len(df))
df_muestra = missing_matrix.sample(n=muestra_size, random_state=42)


# Generamos el mapa de agrupamiento de datos nulos
sns.clustermap(df_muestra,
               cmap='RdYlBu_r', # Restauramos el mapa de colores para mejor visualización
               cbar=True,
               cbar_kws={'label': 'No Nulo (0) vs Nulo (1)'},
               col_cluster=False,      # NO Agrupa variables similares
               row_cluster=False,     # NO Agrupa registros individuales
               xticklabels=True,
               yticklabels=False)

plt.suptitle('Patrón de Nulos sin Agrupamiento (Solo Matriz de Nulos)',
             fontsize=13, fontweight='bold', y=1.02)
plt.show()
```

#### 3.2.2. Análisis de Representatividad (Magnitud Porcentual)

Tras estandarizar los datos, ahora medimos qué proporción de nuestra base de datos está comprometida. Esto define si una variable sigue siendo útil para el modelo o si el ruido es demasiado alto.

```python
# 3.2.2. Análisis de Datos Faltantes (NaN) en Porcentajes

# 1. Calculamos el porcentaje directamente
# Asumimos que ya convertiste 'unknown' a NaN previamente
missing_pct = (df.isnull().sum() / len(df)) * 100

# 2. Filtramos solo los que tienen faltantes y ordenamos
missing_pct = missing_pct[missing_pct > 0].sort_values(ascending=False)

# 3. Configuración del gráfico
plt.figure(figsize=(11, 6))
# Graficamos los valores del porcentaje en el eje X
ax = sns.barplot(x=missing_pct.values, y=missing_pct.index, hue=missing_pct.index, palette='magma', legend=False)

# 4. Ajustes estéticos para "Storytelling"
plt.title('Impacto de Datos Faltantes por Variable (%)', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Porcentaje de valores nulos sobre el total (%)', fontsize=12)
plt.ylabel('Variables con inconsistencias', fontsize=12)
plt.xlim(0, 100)  # Forzamos la escala a 100 para dar perspectiva de "qué tan vacío está"

# 5. Añadir etiquetas de porcentaje al final de las barras
for i, v in enumerate(missing_pct.values):
    # El texto se posiciona dinámicamente según el valor v
    ax.text(v + 1.5, i, f'{v:.2f}%', color='black', va='center', fontweight='semibold')

plt.grid(axis='x', linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()
```

#### 3.2.3. Estrategia de Manejo de Valores Faltantes (NaN)

Tratamiento Ético de Datos Ausentes

Acción: De acuerdo con nuestro marco ético, hemos establecido que los valores unknown no deben ser eliminados a conveniencia ni inventados mediante estimaciones. Para preservar la integridad del análisis, trataremos la ausencia de datos como una categoría informativa.

En la práctica, esto significa que convertiremos los valores NaN en una nueva etiqueta llamada missing. Al hacer esto, le enseñamos al modelo que la falta de respuesta constituye un comportamiento del cliente (por ejemplo, un cliente que decide no revelar su estado laboral) y evitamos sesgar los resultados al inferir datos erróneos.

```python
# --- FASE 1: IDENTIFICACIÓN Y REEMPLAZO ---

# Identificamos de forma dinámica las columnas categóricas que aún contienen valores nulos (NaN)
categorical_cols_with_nan = df[categorical_cols].columns[df[categorical_cols].isnull().any()].tolist()

# Rellenamos los valores nulos con la cadena de texto 'missing'
for col in categorical_cols_with_nan:
    df[col] = df[col].fillna('missing')

# --- FASE 2: VERIFICACIÓN DEL PROCESO ---

print("--- REPORTE DE CONTROL: VALORES FALTANTES (NaN) ---")
# Comprobamos que el reemplazo se haya ejecutado correctamente
print(df[categorical_cols].isnull().sum())

print("\n--- CONCLUSIÓN TÉCNICA ---")
print("Se han rellenado los valores 'unknown' (ahora NaN) con la etiqueta 'missing' en las columnas categóricas especificadas.")
```

Hemos rescatado información que, de otro modo, se habría perdido. Por ejemplo, en la variable `default`, ahora podemos observar si el grupo de clientes que prefiere no declarar su historial crediticio presenta una mayor o menor propensión a contratar un depósito a plazo en comparación con aquellos que sí informan su situación financiera. De esta manera, un vacío de información se transforma en un segmento potencialmente accionable para el Banco Financiero Global.


**Impacto del Tratamiento de Datos Faltantes**

Al convertir los datos faltantes en una categoría formal denominada `missing`, logramos dos objetivos fundamentales para el negocio:

- **Transparencia Ética:**  
  No inventamos ni asumimos información financiera o personal sobre los clientes.

- **Nuevos Insights Estratégicos:**

#### 3.2.4. Manejo de Variables Irrelevantes (Fuga de Datos y Ética)

De acuerdo con los lineamientos de nuestro caso, la variable duration (duración de la última llamada) representa una "trampa analítica". En términos técnicos de Ciencia de Datos, incluir esta variable crearía lo que se conoce como Data Leakage o fuga de datos.

**¿Por qué ocurre esto?**

Para saber cuánto durará una llamada, primero debemos realizarla y finalizarla. En este punto, el cliente ya ha tomado su decisión de suscribirse o no. Si entrenamos nuestro modelo con una variable que el banco no puede conocer antes de levantar el teléfono, el modelo será inútil en la práctica, pues no sirve para predecir, sino para describir algo que ya pasó.

**Justificación Estratégica**

Existe una correlación directa, pero engañosa, entre la duración de una llamada y el resultado final de la campaña. Las conversaciones que terminan en una suscripción exitosa suelen extenderse más tiempo debido al proceso de explicación del producto, resolución de dudas y cierre comercial. En cambio, una respuesta negativa normalmente ocurre en los primeros minutos de la llamada.

Si esta variable se mantuviera dentro del análisis, el algoritmo podría interpretar erróneamente que el éxito depende únicamente de prolongar la conversación. Esto generaría una conclusión artificial y poco útil para el Banco Financiero Global, ya que la duración de la llamada no es una causa del éxito, sino una consecuencia de que el cliente ya mostró interés en contratar el depósito.

**Toma de Decisión: Eliminación Preventiva**

Al eliminar la variable `duration` y generar el dataset `df_clean`, garantizamos que el análisis y los futuros modelos predictivos sean:

- **Realistas:**  
  Basados exclusivamente en información disponible antes del contacto comercial, como datos demográficos, antecedentes socioeconómicos e historial previo del cliente.

- **Éticos:**  
  Cumpliendo con principios de integridad y transparencia al evitar resultados artificialmente elevados provocados por fuga de información (*data leakage*).

- **Accionables:**  
  Entregando una herramienta de segmentación útil para que los ejecutivos del banco puedan priorizar estratégicamente sus listas de llamadas y enfocar esfuerzos en clientes con mayor probabilidad de suscripción.

```python
# --- PREPARACIÓN FINAL ANTES DE LA LIMPIEZA ---

# Convertimos la variable objetivo a numérica para permitir el análisis de correlación en la siguiente sección
df['y_numeric'] = df['y'].apply(lambda x: 1 if x == 'yes' else 0)

# --- ELIMINACIÓN DE LA VARIABLE (ACCIÓN ÉTICA) ---

# Creamos una copia limpia excluyendo 'duration'
# Usamos df_clean de aquí en adelante para asegurar que el análisis sea realista
df_clean = df.drop(columns=['duration'])

print("--- REPORTE DE LIMPIEZA Y TRANSFORMACIÓN ---")
print(f"1. Variable 'duration' eliminada para evitar Data Leakage.")
print(f"2. Variable objetivo 'y' codificada como 'y_numeric' (0/1).")
print("-" * 45)
print(f"Total de variables iniciales: {df.shape[1]}")
print(f"Total de variables para el análisis (df_clean): {df_clean.shape[1]}")
```

Con este último paso, hemos finalizado la fase de curaduría de datos. El dataset ha pasado de ser una base de información bruta, con incertidumbres (`unknown`) y riesgos metodológicos asociados a variables como `duration`, a convertirse en un activo de información limpio, estandarizado y alineado con principios éticos de análisis de datos.

Ahora, el Banco Financiero Global cuenta con una base sólida y confiable para avanzar hacia las siguientes etapas del proyecto, permitiendo desarrollar análisis y modelos predictivos que reflejen escenarios reales y útiles para la toma de decisiones comerciales.

#### 3.2.5. Codificación de Variables Categóricas (Pre-EDA)

Los algoritmos de análisis de datos y *machine learning* no comprenden textos o etiquetas como `"sí"` o `"no"`; necesitan números para calcular correlaciones, distancias y patrones. En este apartado, traduciremos nuestras variables categóricas más importantes a valores numéricos simples.

A este proceso se le conoce como **codificación** o *encoding*. Convertimos variables:

- **Binarias** (ej. tener o no un préstamo)
- **Categóricas** (ej. tipo de contacto)

en valores numéricos (0 y 1, o escalas de valor).

De esta forma, dejamos el terreno preparado para generar mapas de calor y analizar correlaciones en las siguientes etapas del análisis.

```python
from sklearn.preprocessing import LabelEncoder

# --- FASE 1: PREPARACIÓN DEL SET CODIFICADO ---
# Creamos una copia del dataset limpio para no alterar los datos originales
df_encoded = df_clean.copy()

# --- FASE 2: TRANSFORMACIÓN DE CATEGORÍAS A NÚMEROS ---

# Codificamos la variable objetivo 'y'
# 0 representa 'no' (no acepta el depósito) y 1 representa 'yes' (sí lo acepta)
le = LabelEncoder()
df_encoded['y'] = le.fit_transform(df_encoded['y'])

# Creamos diccionarios para asignar un valor numérico lógico a cada categoría
contact_map = {'telephone': 0, 'cellular': 1}
poutcome_map = {'failure': 0, 'nonexistent': 1, 'success': 2}
default_map = {'no': 0, 'yes': 1, 'missing': 2}
housing_map = {'no': 0, 'yes': 1, 'missing': 2}
loan_map = {'no': 0, 'yes': 1, 'missing': 2}

# Aplicamos los mapeos a nuestro nuevo DataFrame
df_encoded['contact'] = df_encoded['contact'].map(contact_map)
df_encoded['poutcome'] = df_encoded['poutcome'].map(poutcome_map)
df_encoded['default'] = df_encoded['default'].map(default_map)
df_encoded['housing'] = df_encoded['housing'].map(housing_map)
df_encoded['loan'] = df_encoded['loan'].map(loan_map)

# --- FASE 3: AUDITORÍA DE LA TRANSFORMACIÓN ---
print("--- MUESTRA DE DATOS CODIFICADOS ---")
display(df_encoded[['contact', 'poutcome', 'default', 'housing', 'loan', 'y']].head())
```

- **Estandarización del Objetivo:**  
  La variable `y` ahora es binaria (0 y 1). Este es un requisito indispensable para que las herramientas matemáticas puedan detectar los perfiles de éxito de los clientes.

- **Jerarquía en Resultados Pasados:**  
  Al mapear `poutcome` (resultado de la campaña anterior), establecimos una escala del 0 al 2, donde 0 representa fracaso y 2 representa éxito. Esta progresión numérica permite medir cuantitativamente la lealtad y el interés histórico del cliente.

Con la ejecución de esta fase, hemos transformado un conjunto de datos bruto y con ruidos en un activo de información estratégico para el Banco Financiero Global. No solo hemos limpiado datos; hemos tomado decisiones basadas en la ética y la eficacia operativa.

El objetivo es identificar qué variables aportan mayor valor estratégico para futuras campañas del Banco Financiero Global.

```python
# ============================================
# PREPARACIÓN DE DATOS
# ============================================

numeric_features = ['age', 'campaign', 'pdays', 'previous',
                    'emp.var.rate', 'cons.price.idx', 'cons.conf.idx',
                    'euribor3m', 'nr.employed']

df_corr = df.copy()
df_corr['y_num'] = df_corr['y'].map({'yes': 1, 'no': 0})
corr_matrix = df_corr[numeric_features + ['y_num']].corr()

# ============================================
# HEATMAP DE CORRELACIÓN
# ============================================

plt.figure(figsize=(14, 12))

# Máscara para triángulo superior (mostrar solo inferior)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

sns.heatmap(corr_matrix, mask=mask, annot=True, fmt=".2f",
            cmap='RdBu_r', center=0, linewidths=0.8,
            cbar_kws={"shrink": 0.7, "label": "Coeficiente de Correlación"},
            annot_kws={"size": 10}, square=True)

plt.title("Mapa de Calor de Correlación entre Variables Numéricas\n" +
          "Valores cercanos a +1 (azul) = relación positiva directa | " +
          "Cercanos a -1 (rojo) = relación inversa",
          fontsize=16, fontweight='bold', pad=20)

plt.xticks(rotation=45, ha='right', fontsize=11)
plt.yticks(rotation=0, fontsize=11)

plt.tight_layout()
plt.show()
```

### CONCLUSIÓN HEATMAP

El análisis de correlación muestra que las variables macroeconómicas son las más relacionadas con la suscripción del depósito a plazo.

Particularmente, `euribor3m`, `nr.employed` y `emp.var.rate` presentan correlaciones negativas moderadas con la variable objetivo, lo que sugiere que los clientes tienden a contratar más depósitos en contextos económicos menos favorables o de mayor incertidumbre.

#### 4.2.1. Análisis Detallado de Distribuciones Numéricas vs. Suscripción: El "Cuándo" y el "Contexto"

Esta subsección profundiza en las variables numéricas más correlacionadas con la suscripción (`y`), mostrando cómo sus distribuciones varían entre los grupos `yes` y `no` mediante **Gráficos de Caja (Boxplots)** y **Gráficos de Densidad (KDE Plots)**. El objetivo es identificar el "cuándo" y el "contexto" en que un cliente es más propenso a contratar el producto.

```python
# ============================================
# GRÁFICOS INDIVIDUALES: Distribución vs "y"
# ============================================

import seaborn as sns
import matplotlib.pyplot as plt

# Seleccionar top 4 variables
corr_with_y = corr_matrix['y_num'].drop('y_num').abs().sort_values(ascending=False)
top_4_features = corr_with_y.head(4).index.tolist()

# Crear figura con subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

colors = {'yes': '#1f77b4', 'no': '#ff7f0e'}

for i, var in enumerate(top_4_features):
    ax = axes[i]

    if var == 'pdays':
        # Filtrar pdays = 999 para una mejor visualización del rango inferior
        df_pdays_filtered = df[df['pdays'] != 999].copy()
        sns.boxplot(data=df_pdays_filtered, x='y', y=var, ax=ax, palette=colors)
        ax.set_title(f'Distribución de {var} (excl. 999) según Suscripción (y)', fontsize=12, fontweight='bold')
        ax.text(0.5, 0.95, 'Nota: pdays=999 (sin contacto previo) excluido para mejor visualización.\nConstituye la mayoría de los casos.',
                transform=ax.transAxes, fontsize=9, ha='center', va='top', bbox=dict(boxstyle='round,pad=0.2', fc='yellow', alpha=0.6))
    else:
        # Boxplot (muestra mediana, cuartiles y outliers)
        sns.boxplot(data=df, x='y', y=var, ax=ax, palette=colors)
        ax.set_title(f'Distribución de {var} según Suscripción (y)', fontsize=12, fontweight='bold')

    ax.set_xlabel('Suscripción a depósito (yes = contrató, no = no contrató)')
    ax.set_ylabel(var)

    # Añadir anotación con la correlación
    r_value = corr_matrix.loc[var, 'y_num']
    ax.annotate(f'Correlación con y: {r_value:.2f}', xy=(0.7, 0.9),
                xycoords='axes fraction', fontsize=10, fontweight='bold',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.suptitle('Análisis de Distribución de Variables Numéricas vs Variable Objetivo "y"',
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# ============================================
# GRÁFICOS INDIVIDUALES: KDE superpuestos
# ============================================

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, var in enumerate(top_4_features):
    ax = axes[i]

    # KDE separado por clase
    for target, color in colors.items():
        subset = df[df['y'] == target][var].dropna()
        sns.kdeplot(subset, ax=ax, color=color, label=f'y={target}',
                    fill=True, alpha=0.4)

    ax.set_title(f'Densidad de {var} según Suscripción (y)', fontsize=12, fontweight='bold')
    ax.set_xlabel(var)
    ax.set_ylabel('Densidad')
    ax.legend()

    # Añadir anotación
    r_value = corr_matrix.loc[var, 'y_num']
    ax.annotate(f'Correlación: {r_value:.2f}', xy=(0.7, 0.85),
                xycoords='axes fraction', fontsize=10, fontweight='bold',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.suptitle('Distribución de Densidad: Comparativa entre Quienes Contratan (yes) y Quienes No (no)',
             fontsize=16, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# ============================================
# CONCLUSIONES ESPECÍFICAS POR VARIABLE
# ============================================

print("\n" + "="*70)
print("CONCLUSIONES: DISTRIBUCIÓN DE VARIABLES vs OBJETIVO 'y'")
print("="*70)

for var in top_4_features:
    r_value = corr_matrix.loc[var, 'y_num']
    print(f"\n📊 {var.upper()} (r = {r_value:+.2f}):")

    # Calcular medianas por grupo
    median_yes = df[df['y'] == 'yes'][var].median()
    median_no = df[df['y'] == 'no'][var].median()

    print(f"   • Mediana para 'yes': {median_yes:.2f}")
    print(f"   • Mediana para 'no': {median_no:.2f}")
    print(f"   • Diferencia: {median_yes - median_no:+.2f}")

    if r_value > 0:
        print(f"   → Los clientes que contratan tienden a tener valores MÁS ALTOS de {var}")
    else:
        print(f"   → Los clientes que contratan tienden a tener valores MÁS BAJOS de {var}")
```

Los gráficos de caja (*boxplots*) y densidad (*KDE plots*) a continuación nos ofrecen una vista granular de cómo las cuatro variables numéricas más correlacionadas con la suscripción (`nr.employed`, `pdays`, `euribor3m`, `emp.var.rate`) se distribuyen entre los clientes que `sí` (azul) y `no` (naranja) aceptaron el depósito. ### 4.3. Análisis de Variables Categóricas: "El Perfil del Cliente"

En marketing analítico, las decisiones más valiosas no nacen únicamente de gráficos atractivos o intuiciones visuales. Para construir una campaña realmente efectiva, necesitamos identificar, con evidencia estadística, qué factores influyen de forma significativa en la decisión del cliente.

El desafío es que gran parte de nuestro dataset está compuesto por variables categóricas, como el tipo de trabajo (job), el mes de contacto (month) o el resultado de campañas anteriores (poutcome). Estas variables no pueden evaluarse mediante correlaciones tradicionales como Pearson, ya que no representan valores numéricos continuos.

Para resolver este problema, aplicamos el Coeficiente V de Cramer, una métrica especializada en medir la fuerza de asociación entre variables categóricas.

¿Qué nos permite descubrir esta técnica?
Medir influencia real:
Cuantificar qué tan asociada está cada variable con la decisión final de suscripción (y).
Separar señales del ruido:
Detectar cuáles atributos aportan valor predictivo y cuáles tienen un impacto mínimo o irrelevante.
Priorizar la estrategia comercial:
Dirigir los esfuerzos del banco hacia los factores que muestran mayor capacidad de influir en la conversión del cliente.

En otras palabras, esta etapa transforma el análisis exploratorio en inteligencia accionable. Ya no observamos únicamente tendencias; comenzamos a identificar los verdaderos motores estadísticos detrás del éxito de una campaña.

```python
import scipy.stats as ss
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Función experta para calcular Cramer's V con corrección de sesgo
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    if min(k-1, r-1) == 0:
        return 0.0
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# 1. Selección de variables categóricas (usamos el df original o df_clean para textos)
categorical_features = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week', 'poutcome']

# 2. Cálculo de resultados
cramers_v_results = {feat: cramers_v(df[feat], df['y']) for feat in categorical_features}
cramers_v_series = pd.Series(cramers_v_results).sort_values(ascending=False)

# 3. Visualización del Mapa de Calor (Heatmap)
plt.figure(figsize=(7, 8))
sns.heatmap(
    cramers_v_series.to_frame(name="Fuerza de Asociación"),
    annot=True,
    cmap='magma', # Usamos magma para mantener la consistencia estética de tu proyecto
    fmt=".2f",
    linewidths=.8,
    cbar_kws={'label': "Índice Cramer's V"}
)

plt.title("Influencia de Categorías en la Suscripción", fontsize=15, pad=20, fontweight='bold')
plt.ylabel('Atributos del Cliente', fontsize=12)
plt.xlabel('')
plt.xticks([])
plt.tight_layout()
plt.show()
```

**Hallazgos Estratégicos: Los Verdaderos Motores de la Decisión**

El análisis de asociación mediante el **Coeficiente V de Cramer** nos permitió separar el “ruido” estadístico de las señales realmente relevantes para el negocio. Más allá de simples conteos o porcentajes, este enfoque revela qué factores tienen mayor capacidad para inclinar la decisión del cliente hacia la contratación.


**1. El Peso del Historial (`poutcome`)**

El hallazgo más contundente del análisis es que el resultado de la campaña anterior (`poutcome`) presenta la asociación más fuerte con el éxito de la campaña actual.

**Lectura Estratégica**

Un cliente que ya aceptó un producto financiero anteriormente posee una probabilidad significativamente mayor de volver a contratar. Esto confirma que la confianza previa y la experiencia positiva generan un efecto acumulativo sobre futuras conversiones.

> En términos prácticos: el éxito pasado se convierte en uno de los mejores predictores del éxito futuro.

**Lectura Estratégica**

> El momento correcto puede ser tan importante como la oferta correcta.


**3. Perfil del Cliente vs. Contexto Comercial**

Uno de los hallazgos más interesantes del análisis es que variables demográficas tradicionales como `housing`, `loan` o `marital` muestran niveles de asociación considerablemente más bajos.

**Lectura Estratégica**

### 5.3. Codificación de Variables Categóricas

Los algoritmos de machine learning operan mejor con datos numéricos. Por lo tanto, en esta subsección, transformaremos nuestras variables categóricas a un formato numérico. Utilizaremos la codificación ordinal para variables como 'education', que tienen un orden lógico, y One-Hot Encoding para variables nominales, creando nuevas columnas binarias que representan cada categoría de forma independiente. Esto prepara el terreno para que los modelos puedan interpretar correctamente estas características.

```python
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

# --- 5.3.1. Codificación Ordinal para 'education' (Si aún existe) ---
# Asegúrate de que education_order incluye todos los valores únicos, incluyendo 'missing' si es posible que aparezca.
education_order = [
    'illiterate', 'basic.4y', 'basic.6y', 'basic.9y', 'high.school',
    'professional.course', 'university.degree', 'missing', 'unknown' # Añadir 'unknown' por si acaso
]

# Crear y aplicar el OrdinalEncoder solo si 'education' está presente
if 'education' in X.columns and len(categorical_ordinal_features) > 0:
    ordinal_encoder = OrdinalEncoder(categories=[education_order], handle_unknown='use_encoded_value', unknown_value=-1)
    X['education_encoded'] = ordinal_encoder.fit_transform(X[['education']])
    X = X.drop('education', axis=1)
    print("Variable 'education' codificada ordinalmente.")
    display(X[['education_encoded']].head())
else:
    print("La variable 'education' ya ha sido procesada o no existe para codificación ordinal.")

# --- 5.3.2. One-Hot Encoding para variables nominales ---
if len(categorical_nominal_features) > 0:
    onehot_encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
    encoded_nominal_features = onehot_encoder.fit_transform(X[categorical_nominal_features])

    # Convertir las características codificadas a un DataFrame
    encoded_nominal_df = pd.DataFrame(encoded_nominal_features,
                                      columns=onehot_encoder.get_feature_names_out(categorical_nominal_features),
                                      index=X.index)

    # Unir las características codificadas con el DataFrame original y eliminar las columnas originales
    X = pd.concat([X.drop(columns=categorical_nominal_features), encoded_nominal_df], axis=1)
    print("\nVariables categóricas nominales codificadas mediante One-Hot Encoding.")
    display(X.head())
else:
    print("\nNo hay variables categóricas nominales pendientes de One-Hot Encoding o ya fueron procesadas.")
```

### 5.4. Escalado de Variables Numéricas

Las variables numéricas con diferentes rangos de valores pueden sesgar los algoritmos de machine learning, especialmente aquellos basados en distancias. En esta sección, aplicaremos un proceso de escalado para normalizar nuestras variables numéricas, asegurando que todas contribuyan de manera equitativa al modelo. Esto mejora el rendimiento y la estabilidad de muchos algoritmos de aprendizaje automático.

```python
from sklearn.preprocessing import StandardScaler

# Aplicar StandardScaler a las variables numéricas
scaler = StandardScaler()
X_scaled_numerical = scaler.fit_transform(X[numerical_features])

# Convertir a DataFrame
X_scaled_numerical_df = pd.DataFrame(X_scaled_numerical,
                                     columns=numerical_features,
                                     index=X.index)

# Unir las características escaladas con el resto de las características procesadas
X_processed = pd.concat([X_scaled_numerical_df, X.drop(columns=numerical_features)], axis=1)

print("Variables numéricas escaladas usando StandardScaler.")
display(X_processed.head())
print(f"Dimensiones finales de X_processed: {X_processed.shape}")
```

### Visualización del Escalado de Variables Numéricas
